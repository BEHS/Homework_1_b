```python
huntsville_location = (34.72806, -86.64733)
data = {"Limestone Creek": [34.75166667, -86.82305556], 
        "Big Wills Creek": [34.67, -87.76722222], 
        "Big Nace Creek":[34.67, -87.31722222], 
        "Flint River":[34.74916667, -86.44666667],
        "Pinhook Creek":[34.73027778,-86.5933333],
        "Tennessee River at Decature":[34.61138889, -86.97527778],
        "Tennessee River at Florence":[34.78694444, -87.67],
        "Aldridge Creek":[34.62388889, -86.54111111],
        "Tennessee River below Guntersville Dam":[34.42138889, -86.39388889],
        'Tennessee River above Guntersville Dam':[34.4217, -86.3927],
        "Indian Creek": [34.69722222, -86.69972222],
        "Bear Creek": [34.44388889, -88.11555556],
        "Tennessee River near Scottsboro":[34.64361111, -85.98027778],
        "Aldridge Creek at Sherwood Drive in Huntsville": [34.66416667, -86.55416667],
        "Aldridge Creek at Toney Drive in Huntsville": [34.69805556, -86.55055556],
        "Paint Rock River": [34.62416667, -86.30638889],
        "Tennessee River at Whitesburg": [34.57166667, -86.55805556]
}
m = folium.Map(location=huntsville_location, zoom_start=9)
for name, coords in data.items():
    folium.Marker(
        location=coords,
        popup=name,
        tooltip="Click me!",
        icon=folium.Icon(icon="cloud")
        ).add_to(m)

m
```
